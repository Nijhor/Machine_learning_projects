```python
import pandas as pd 
import requests 
from bs4 import BeautifulSoup 
import numpy as np 
```

```python
url = 'https://www.imdb.com/list/ls050782187/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup)
```

```python
m_name = []
m_year = []
m_gross = []
m_rateing = []
m_vote = []
m_catagory = []
```

```python
movie_data = soup.findAll('div',attrs={'class': 'lister-item mode-detail'})
```

```python
for i in movie_data:
  name = i.h3.a.text
  m_name.append(name)

  year = i.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
  m_year.append(year)

  num = i.find_all('span', attrs = {'name': 'nv'})
  vote = num[0].text
  m_vote.append(vote)

  gross = num[1].text if len(num)>1 else '*****'
  m_gross.append(gross)

  rate = i.find('span', class_ = 'ipl-rating-star__rating').text
  m_rateing.append(rate)

  category = i.find('span', class_='genre').text.replace('\n','')
  m_catagory.append(category)


```

```python
df_movie = pd.DataFrame({'Name':m_name,'Catagory':m_catagory,'Year':m_year,'Rating':m_rateing,'Gross_collection':m_gross,'Voting':m_vote})
```
